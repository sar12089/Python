```python
from functools import reduce
import math
import random
a=dict()
stat={1:'sample',0:'population'}
def sumdat(dat):
    a[dat+'_sum'] = reduce(lambda x,y:x+y,a[dat])
    print("\tSum: ",a[dat+'_sum'])
def mean(dat):
    """This function gets Key (dictionary) as a parameter and returns back the mean of the values stored in that Key"""
    a[dat+'_mean'] = reduce(lambda x,y:x+y,a[dat])/len(a[dat])
    if float(a[dat+'_mean'])==int(a[dat+'_mean']):
        a[dat+'_mean']=int(a[dat+'_mean'])
    print("\tMean: ",a[dat+'_mean'])
def median(dat):
    """This function gets Key (dictionary) as a parameter and returns back the median of the values stored in that Key"""
    mod=len(a[dat])%2
    if mod==1:
        a[dat+'_median']=a[dat][int(len(a[dat])/2)]
    else:
        med=int(len(a[dat])/2)
        a[dat+'_median']=(a[dat][med-1]+a[dat][med])/2
    if float(a[dat+'_median'])==int(a[dat+'_median']):
        a[dat+'_median']=int(a[dat+'_median'])
    print("\tMedian:", a[dat+'_median'])
def rangedat(dat):
    """This function gets Key (dictionary) as a parameter and returns back the range of the values stored in that Key"""
    a[dat+'_range']=a[dat][len(a[dat])-1]-a[dat][0]
    a[dat+'_max']=a[dat][len(a[dat])-1]
    a[dat+'_min']=a[dat][0]
    print("\tMin: ",a[dat+'_min'])
    print("\tMax: ",a[dat+'_max'])
    print("\tRange: ",a[dat+'_range'])
def variance(dat,stat_in):
    """This function gets Key (dictionary) as a parameter and returns back the variance of the values stored in that Key. Mean will be calculated by intern."""
    if dat+'_mean' not in a:
        mean(dat)
    var=0
    for i in a[dat]:
        var=var+pow(i-a[dat+'_mean'],2)
    a[dat+'_variance']=var/(len(a[dat])-stat_in)
    print("\tVariance: ",a[dat+'_variance'])
def SD(dat):
    """This function gets Key (dictionary) as a parameter and returns back the SD of the values stored in that Key. Mean and variance will be calculated by intern."""
    if dat+'_variance' not in a:
        variance(dat,stat_in)
    a[dat+'_SD']=math.sqrt(a[dat+'_variance'])
    print("\tStandard Deviation: ",a[dat+'_SD'])
def percentiles(p,dat):
    per=p*len(a[dat])
    if type(per).__name__=='float' and int(per)!=per:
        per=int(per)
        val=a[dat][per]
    else:
        per=int(per)
        val=(a[dat][per-1]+a[dat][per])/2
    if float(val)==int(val):
        val=int(val)
    return val
def quartiles(dat):
    """This function gets Key (dictionary) as a parameter and returns back the Quartiles (Q1,Q3) of the values stored in that Key"""
    a[dat+'_Q1']=percentiles(0.25,dat)
    a[dat+'_Q3']=percentiles(0.75,dat)
def IQR(dat):
    """This function gets Key (dictionary) as a parameter and returns back the IQR and 1.5IQR of the values stored in that Key. Quartiles will be calculated by intern."""
    if dat+'_Q3' not in a:
        quartiles(dat)
    a[dat+'_IQR']=a[dat+'_Q3']-a[dat+'_Q1']
    a[dat+'_1.5IQR']=1.5*a[dat+'_IQR']
    print("\tQuartile Q1: ",a[dat+'_Q1'])
    print("\tQuartile Q3: ",a[dat+'_Q3'])
    print("\tIQR: ",a[dat+'_IQR'])
    print("\t1.5 IQR: ",a[dat+'_1.5IQR'])
def coeffvar(dat):
    """This function gets Key (dictionary) as a parameter and returns back the coefficient of variation of the values stored in that Key. Mean, variance, SD will be calculated by intern."""
    if dat+'_SD' not in a:
        SD(dat)
    a[dat+'_coeffvariation']=round((a[dat+'_SD']/a[dat+'_mean'])*100,6)
    print("\tCoefficient of Variation",a[dat+'_coeffvariation'])
def limit(dat):
    """This function gets Key (dictionary) as a parameter and returns back the upper limit and lower limit of the values stored in that Key. Quartiles, IQR will be calculated by intern."""
    if dat+'_Q3' not in a:
        quartiles(dat)
    if dat+'_IQR' not in a:
        IQR(dat)
    a[dat+'_lowerlimit']=a[dat+'_Q1']-a[dat+'_1.5IQR']
    a[dat+'_upperlimit']=a[dat+'_Q3']+a[dat+'_1.5IQR']
    print("\tLower Limit: ",a[dat+'_lowerlimit'])
    print("\tUpper Limit: ",a[dat+'_upperlimit'])
def all_basics(dat,d,stat_in):
    a[dat+'_unsort']=d
    #Sort the data
    a[dat]=sorted(d)
    print("Dataset: ",a[dat])
    sumdat(dat)
    mean(dat)
    median(dat)
    rangedat(dat)
    variance(dat,stat_in)
    SD(dat)
    quartiles(dat)
    coeffvar(dat)
    limit(dat)
def covcorel(s,t,stat_in):
    covnum=0
    #for j in range(min(len(a[s]),len(a[t]))):
    #    covnum=covnum+((a[s+'_unsort'][j]-a[s+'_mean'])*(a[t+'_unsort'][j]-a[t+'_mean']))
    #    cov=covnum/(min(len(a[s]),len(a[t]))-stat_in)
    for i,j in zip(a[s+'_unsort'],a[t+'_unsort']):
        covnum=covnum+(i-a[s+'_mean'])*(j-a[t+'_mean'])
        cov=covnum/(min(len(a[s]),len(a[t]))-stat_in)
    print("Covariance of {0} and {1}: {2}".format(s,t,cov))
    cocov=cov/(a[s+'_SD']*a[t+'_SD'])
    print("Correlation coefficient between {0} and {1}: {2}".format(s,t,cocov))
def randcheck(dat,numofsamp):
    samp=[]
    #print("---")
    for items in range(0,numofsamp):
        samp.insert(items,a[dat][random.randint(0,len(a[dat])-1)])
    #samp=set(samp)
    #samp=list(samp)
    #print("---",samp,len(samp),numofsamp)
    #if len(samp)<numofsamp:
        #print(samp)
        #samp=randcheck(dat,numofsamp)
    #samp=list(samp)
    return samp
def samplegen(dat):
    sampsize=int(input("What is the sample size? - "))
    numofsamp=int(input("Number of data in a sample - "))
    sampmean=[]
    sampvar=[]
    st=set(a[dat])
    if len(st)>=numofsamp+1:
        for sz in range(1,sampsize+1):
            #samp=[]
            #for items in range(0,numofsamp):
                #samp.insert(items,a[dat][random.randint(0,len(a[dat])-1)])
            sampname=dat+'_sample_'+str(sz)
            a[sampname]=randcheck(dat,numofsamp)
            print("{0}: {1}".format(sz,a[sampname]))
            mean(sampname)
            sampmean.insert(sz-1,a[sampname+'_mean'])
            sampvar.insert(sz-1,pow(a[sampname+'_mean']-a[dat+'_mean'],2))
            print("The mean of {0} is {1}".format(sampname,a[sampname+'_mean']))
            print("The variance of {0} mean from population mean is {1}".format(sampname,pow(a[sampname+'_mean']-a[dat+'_mean'],2)))
        a[dat+'_samples_mean']=sampmean
        a[dat+'_samples_variance']=sampvar
        print(a[dat+'_samples_mean'])
        mean(dat+'_samples_mean')
        mean(dat+'_samples_variance')
        if a[dat+'_samples_mean_mean']==a[dat+'_mean']:
            print("Mean of the samples mean {0} is same as population mean {1}".format(a[dat+'_samples_mean_mean'],a[dat+'_mean']))
        else:
            print("Mean of the samples mean {0} is NOT same as population mean {1}".format(a[dat+'_samples_mean_mean'],a[dat+'_mean']))
        a[dat+'_samples_mean_SD']=math.sqrt(a[dat+'_samples_variance_mean'])
        a[dat+'_CLT']=a[dat+'_SD']/a[dat+'_samples_mean_SD']
        print(a[dat+'_SD'])
        print(a[dat+'_samples_mean_SD'])
        if a[dat+'_CLT']==math.sqrt(numofsamp):
            print("The Central limit theorem is proved - SD of population/SD of samples mean {0} is equal to square root of number of data in samples {1}".format(a[dat+'_CLT'],math.sqrt(numofsamp)))
        else:
            print("The Central limit theorem is NOT proved - {0} is not equal to {1}".format(a[dat+'_CLT'],math.sqrt(numofsamp)))
        a[dat+'_CLT']=a[dat+'_SD']/math.sqrt(numofsamp)
        print("Central Limit Theorem - Population SD/SQRT of number of samples -",a[dat+'_CLT'])
    else:
        print("Data is not enough to take sample")

if __name__ == "__main__":
    ar_count = int(input("Enter the number of columns of data: "))
    stat_in=int(input("Enter the type of the data:\n\t0 - Population\n\t1 - Sample\t - "))
    dats=[]
    for i in range(ar_count):
        dd=input("Enter the dataset name (Key): ".format(i))
        dats.insert(i,dd)
        a["Datasetname"]=dats
        b=list(map(int, input("Enter the values separated by SPACE: ").rstrip().split()))
        all_basics(dd,b,stat_in)
    if ar_count>1:
        cc=input("Do you want to find relation between two attributes? - Y or N - ")
        if cc=='Y':
            st = input("Enter the two dataset names to find covariance & correlation ").split()
            s = st[0]
            t = st[1]
            covcorel(s,t,stat_in)
    if stat_in==0:
        q=input("Do you want to generate samples? - Y or N - ")
        if q=='Y':
            samp=input("For which column you want the samples to be generated? - ").rstrip()
            samplegen(samp)
```
