```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
```

```python
train['Source']='train'
test['Source']='test'
```

```python
df=pd.concat([train,test],axis=0,sort=False,ignore_index=True)
```

```python
df.head()
```

```python
for x in df.columns:
    if df[x].sum()==0:
        print(x)
        df.drop(columns=x, axis=1, inplace=True)
```

```python
 df.duplicated().sum()
```

```python
X=[x for x in df.columns if x not in (['ID','Source','TARGET','Target_Class'])]
X_train=df[X]
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
df_mod = sc_X.fit_transform(X_train)
df_mod=pd.DataFrame(df_mod,columns=X)
df_mod=pd.concat([df_mod,df['ID']],axis=1)
df_mod=pd.merge(df_mod,df[['ID','Target_Class','Source','TARGET']],how='left')
```

```python
df_mod.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
n_components=60
pca = PCA(n_components)
cols=[]
for i in range(n_components):
    nm='PCA'+str(i+1)
    cols.append(nm)
```

```python
print(cols)
```

```python
X=[x for x in df_mod.columns if x not in (['ID','Source','TARGET','Target_Class'])]
```

```python
principalComponents = pca.fit_transform(df_mod[X])
```

```python
PCAdf = pd.DataFrame(data = principalComponents
             , columns = cols)
```

```python
PCAdf.shape
```

```python
PCAdf=pd.concat([PCAdf,df_mod[['Target_Class','Source','ID']]],axis=1)
```
