```python
dataf.info()
```

```python
data_column_category=dataf.select_dtypes(exclude=[np.number]).columns
data_column_category
```

```python
data_column_Integer=dataf.columns.difference(data_column_category)
data_column_Integer
```

```python
dataf[data_column_category].nunique()
```

```python
dataf_one_hot=pd.get_dummies(dataf[data_column_category].drop(columns=['Product_ID','Source']))
```

```python
dataf_one_hot.head()
```

```python
data_model=pd.concat([dataf_one_hot,dataf['Product_ID'],dataf['Source'],dataf[data_column_Integer]],axis=1)
```

```python
data_model.sort_index().head()
```

```python
X=data_model[data_model['Source']=='train'].drop(columns=['Product_ID','Source','Purchase','User_ID'])
Y=data_model[data_model['Source']=='train']['Purchase']
```

```python
import sklearn.feature_selection
select=sklearn.feature_selection.SelectKBest(k=20)
selected_features=select.fit(X,Y)
indices_selected=selected_features.get_support(indices=True)
colnames_selected=[X.columns[i] for i in indices_selected]
colnames_selected
```

```python
train_modified=data_model[data_model['Source']=='train'].drop(columns=['Source'],axis=1)
#train_modified.drop(columns=['index'],inplace=True)
train_modified.shape
```

```python
train_modified.head()
```

```python
test_modified=data_model[data_model['Source']=='test'].drop(columns=['Source','Purchase'],axis=1)
#test_modified.drop(columns=['index'],inplace=True)
test_modified.shape
```

```python
train_modified.to_csv("train_ready_for_model.csv",index=False)
```
